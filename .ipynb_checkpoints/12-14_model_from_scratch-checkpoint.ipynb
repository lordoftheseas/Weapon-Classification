```python
number_classes = 9
input_shape = (144, 144, 1) # images should be 144x144 and gray color format
```

```python
def inspectDatabase(curDatabase):
    print(f"The cardinality of the database is {curDatabase.cardinality()}")
    
    class_names = curDatabase.class_names
    print(f"The database contains {len(class_names)} categories, and they are: {class_names}")
    
    plt.figure(figsize=(10, 10))
    for images, labels in curDatabase.take(1):
        for i in range(9):
            ax = plt.subplot(3, 3, i + 1)
            plt.imshow(images[i].numpy().astype("uint8"))
            plt.title(class_names[labels[i]])
            plt.axis("off")
            if i == 0:
                print(images[i])
```

```python
#data_path = "G:\\My Drive\\University classes\\CSE 455\\12-07_project\\small_dataset_for_training\\"
data_path = "C:\\Users\\Bee\\OneDrive - University at Buffalo\\Fall Semester 2024\\CSE 455\\FP\\Weapon-Classification\\Dataset"

training_data, validation_data = keras.utils.image_dataset_from_directory(
    data_path,
    color_mode = "grayscale",
    image_size = (input_shape[0], input_shape[1]),
    validation_split = 0.1,
    subset = "both",
    seed = 44,
)
```

```python
inspectDatabase(training_data)
```

```python
inspectDatabase(validation_data)
```

```python
model = keras.Sequential(
    [
        keras.layers.Input(shape=input_shape, name="Input"),
        keras.layers.Rescaling(scale=1./255),
        
        # Convolutional Layers
        keras.layers.Conv2D(32, (3, 3), padding="same", activation="relu", name="Conv_1"),
        keras.layers.Conv2D(32, (3, 3), padding="valid", activation="relu", name="Conv_2"),
        keras.layers.MaxPooling2D(pool_size=(2, 2), strides=2, name="MaxPool_1"),
        keras.layers.BatchNormalization(),
        
        keras.layers.Conv2D(64, (3, 3), padding="same", activation="relu", name="Conv_3"),
        keras.layers.Conv2D(64, (3, 3), padding="valid", activation="relu", name="Conv_4"),
        keras.layers.MaxPooling2D(pool_size=(2, 2), strides=2, name="MaxPool_2"),
        keras.layers.BatchNormalization(),
        
        keras.layers.Conv2D(128, (3, 3), padding="same", activation="relu", name="Conv_5"),
        keras.layers.Conv2D(128, (3, 3), padding="valid", activation="relu", name="Conv_6"),
        keras.layers.MaxPooling2D(pool_size=(2, 2), strides=2, name="MaxPool_3"),
        keras.layers.BatchNormalization(),
        
        # Add dropout and Flatten
        keras.layers.Dropout(0.5, name="Dropout_1"),
        keras.layers.Flatten(name="Flatten"),
        
        # Fully connected layers
        keras.layers.Dense(512, activation="relu", name="Dense_1"),
        keras.layers.Dropout(0.5, name="Dropout_2"),
        
        # Output Layer
        keras.layers.Dense(number_classes, activation="softmax", name="Output")
    ]
)
```

```python
model.summary()
#keras.utils.plot_model(model)
```

```python
model.compile(
    optimizer = "adam",
    loss = "sparse_categorical_crossentropy",
    metrics = ["sparse_categorical_accuracy"],
)
```

```python
history = model.fit(
    training_data,
    validation_data = validation_data,
    epochs = 10,
)
```

```python
print(history)
```

```python
# Test the evaluation function

evaluation_results_validation = model.evaluate(
    validation_data,
    verbose = 2,
)
```
